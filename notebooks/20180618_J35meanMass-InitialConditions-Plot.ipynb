```python
from matplotlib.colors import Normalize
import seaborn as sns
import planetePlots

fig, ax = plt.subplots()
norm = matplotlib.colors.Normalize(vmin=-1.5, vmax=1.5)
cmap = sns.diverging_palette(h_neg=250, h_pos=15., s=99, l=50,
                                 n=19, center="dark", as_cmap=True)
sc = ax.scatter(J35stats['disk_solid_initial'], J35stats['theoPlanAcore'], c=J35stats['meanMass_dev2T1'],
            cmap=cmap, norm=norm, edgecolors='white', linewidth=.5)
plt.loglog()
ax.get_xaxis().set_major_formatter(ticker.ScalarFormatter())
ax.get_yaxis().set_major_formatter(ticker.ScalarFormatter())
ax.get_xaxis().set_minor_formatter(ticker.NullFormatter())

cbar = fig.colorbar(sc, ticks=[-1.5, -1., -.5, 0, .5, 1., 1.5], label='Mean Mass Difference [$\mathrm{M_{\oplus}}$]')
cbar.ax.set_yticklabels(['$<$ -1.5','-1.0', '-0.5', '0', '0.5', '1.0', '$>$ 1.5'])  # vertically oriented colorbar

ax.set_xlabel('Initial Solid Mass [$\mathrm{M_{\oplus}}$]')
ax.set_ylabel('Disk Cutoff Radius [au]')
```
